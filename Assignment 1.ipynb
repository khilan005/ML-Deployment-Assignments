# Question 2

Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line.

```python
a=[] 
for x in range(2000, 3201):
    if ( x % 7 == 0 ) and ( x % 5 != 0 ):
        a.append(str(x))
print (','.join(a))
```

# Question 3

Write a Python program to accept the user's first and last name and then getting them printed in
the the reverse order with a space between first name and last name.

```python
#To input first name
first_name = input("Input your First Name : ")

#to input last name
last_name = input("Input your Last Name : ")

#for  Printing Output
print (first_name[::-1] + " " + last_name[::-1])
```

# Question 4

Write a Python program to find the volume of a sphere with diameter 12 cm.

```python
pi =  3.14159265358   # define value of pie
r = float(6)  # radius = daimeter / 2

#calculate the volume of sphere
volume=(4/3)*pi*r*r*r  #volume=(4/3)*pi*r**3

#now printing the output 
print("volume of the sphere will be : " ,volume)
print("volume of the sphere will be :  %.4f" %volume) #answer upto 4 decimal digits
```

# Question 2

```python
n=5; # for maximum number of * in one longest row
for i in range(0,n):   #for increasing order
    for j in range(0,i+1): 
        print ('* ', end="")
    print('')

for i in range(n,0,-1): # for decerasing oreder by one by one value 
    for j in range(0,i-1):
        print('* ', end="")
    print('')
```
